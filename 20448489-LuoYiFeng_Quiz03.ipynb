```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
```

```python
url = 'https://www.hko.gov.hk/en/cis/normal/1981_2010/dnormal10.htm'
source = requests.get(url).text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
#print(soup.prettify())
```

**After carefully observing the webpage code, I found that the webpage contains two tables, and all the contents of these two tables are under the parent node (attribute = table)** 

![a1](./figures/a1.png)

**Because there are two tables in this page, So I used `find_all()` method.To find all the information of the list in a list**

```python
tables = soup.find_all('table')
#tables
```

**Because `find_all()` method return a list of object.So i need to access each tables value individually.**
**SO here it returns a list of table 

### Get the Table1 in the web page

**And within the `tables[0]`, I find all the rows of the table were in the format like below in the Web Html Page**

![a2](./figures/a2.png)

**with the `attribute = tr` and `style = 'border:1px solid;text-align:center;'` 

**So i used `find_all()`method to get all the rows in table one within a list which is called rows1**

- `rows1` stands for all the rows in the table1
- `tables[0]` stands for the first object in the tables list

```python
rows1 = tables[0].find_all('tr', style = 'border:1px solid;text-align:center;')
#rows1
```

**Within the rows1 list, I found the Web Html structure like below**
![a3](./figures/a3.png)
**And it shows that for each values in one row, it has a unique `headers` with the `attribute = td` 
**With means for each for loops, i can access the value individually, and return them in a list**

```python
date_list = []                        # date
press_list = []                       # Mean Pressure
tempmax_list = []                     # Air Temperature Mean Maximum(deg.C)
tempmean_list = []                    # Air Temperature Mean(deg.C)
tempmin_list = []                     # Air Temperature Mean Minimum(deg.C)
wet_list = []                         # Wet Bulb(deg.C)
dew_list = []                         # Dew Point(deg. C)
rh_list = []                          # Relative Humidity(%)
rain_list = []                        # Mean Daily Rainfall(mm)
cloud_list = []                       # Amount of Cloud(%)
```

```python
for row in rows1:
    date = row.find('td', headers='date').text
    press = row.find('td', headers='press').text
    tempmax = row.find('td', headers='temp tmax').text
    tempmean = row.find('td', headers='temp tmean').text
    tempmin = row.find('td', headers='temp tmin').text
    wet = row.find('td', headers='wet').text
    dew = row.find('td', headers='dew').text
    rh = row.find('td', headers='rh').text
    rain = row.find('td', headers='rain').text
    cloud = row.find('td', headers='cloud').text

    
    date_list.append(date)
    press_list.append(press)
    tempmax_list.append(tempmax)
    tempmean_list.append(tempmean)
    tempmin_list.append(tempmin)
    wet_list.append(wet)
    dew_list.append(dew)
    rh_list.append(rh)
    rain_list.append(rain)
    cloud_list.append(cloud)


# print(date_list)
# print(press_list)
# print(tempmax_list)
# print(tempmean_list)
# print(tempmin_list)
# print(wet_list)
# print(dew_list)
# print(rh_list)
# print(rain_list)
# print(cloud_list)
```

**Create a `dataFrame` to store table1** 

```python
table1 = pd.DataFrame({
    'Date': date_list,
    'Mean Pressure': press_list,
    'Air Temperature Mean Maximum(deg.C)': tempmax_list,
    'Air Temperature Mean(deg.C)': tempmean_list,
    'Air Temperature Mean Minimum(deg.C)': tempmin_list,
    'Wet Bulb(deg.C)': wet_list,
    'Dew Point(deg. C)': dew_list,
    'Relative Humidity(%)': rh_list,
    'Mean Daily Rainfall(mm)': rain_list,
    'Amount of Cloud(%)': cloud_list
})
```

```python
for row in rows2:
    sun = row.find('td', headers = 'sun').text
    wind2_dd = row.find('td', headers = 'wind2 dd').text
    wind2_ff = row.find('td', headers = 'wind2 ff').text
    wind_seatemp = row.find('td', headers = 'wind seatemp').text
    wind_seatemp2 = row.find('td', headers = 'wind seatemp2').text
    
    
    
    
    sun_list.append(sun)
    wind2_dd_list.append(wind2_dd)
    wind2_ff_list.append(wind2_ff)
    wind_seatemp_list.append(wind_seatemp)
    wind_seatemp2_list.append(wind_seatemp2)
```
