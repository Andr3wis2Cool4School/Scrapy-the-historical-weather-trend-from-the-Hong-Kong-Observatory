```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
url = 'https://www.hko.gov.hk/tc/cis/normal/1981_2010/dnormal10.htm#'
pages = requests.get(url)
pages
```

```python
pages.content
```

```python
soup = BeautifulSoup(pages.content, 'html.parser')
```

```python
soup
```

```python
warp = soup.find(id = 'wrap')
```

```python
bodyArea = warp.find(class_= 'bodyArea')
```

```python
container = bodyArea.find(class_= 'container')
container
```

```python
contentArea = container.find(class_='contentArea')
contentArea
```

```python
mainContent = contentArea.find(id = 'mainContent')
mainContent
```

```python
border_blank = mainContent.find(class_= 'border_blank')
border_blank
```

```python
div_responsive_table_1 = border_blank.find_all(class_= 'div_responsive_table_1')
div_responsive_table_1[0]
```

```python
contentArea = container.find(class_= 'contentArea').find(id = 'mainContent').find(class_='border_blank')
border_blank
```
